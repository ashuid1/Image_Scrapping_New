```python
import requests
import pandas as pd
import chromedriver_binary

from selenium import webdriver
from bs4 import BeautifulSoup

link = 'https://stock-pictures.netlify.app'
```

```python
driver = webdriver.Chrome()
```

```python
driver.get(link)
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')        ### here page source to get the source of the website.
```

```python
data = []

for sp in soup.find_all('div', class_ = 'container'):

    if('gif' not in sp.find('img').get('src')):
        
        link = sp.find('img').get('src')
        tags = list(set(sp.find('div',class_ = 'tags').text[7:].split(' ')))
        
        likes    = int(sp.find('div',class_ = 'likes-comments').find_all('span')[0].text[:-6])
        comments = int(sp.find('div',class_ = 'likes-comments').find_all('span')[1].text[:-9])
        
        data.append([link, ' '.join(tags), likes, comments])
```

```python
data
```

```python
df = pd.DataFrame(data, columns = ['link','tags','likes','comments'])

df.head()
```

```python
df.isnull().sum()
```

```python
df
```

```python
df.to_csv('Stock_Image.csv', index = False)
```

### To Download the images from the link...

```python
import requests
import pandas as pd
from tqdm import tqdm

df = pd.read_csv('Stock_Image.csv')

df.head()
```

### Downloading the image

```python
link = df['link'][0]

res = requests.get(link)
```

```python
fd = open('Img.png','wb')

fd.write(res.content)               #### here res.content convert the file into binary format then download the image....

fd.close()
```

### Downloading the Total Images in the folder Imgs.

```python
path = []

for link in tqdm(df['link']):
    
    res = requests.get(link)
    
    fd = open('Imgs/'+link.split('/')[-1],'wb')
    fd.write(res.content)
    fd.close()
    
    path.append('Imgs/'+link.split('/')[-1])
```

```python
df['path'] = path      #### here we creating the path of image to access the image which is offline
```

```python
df.head()
```

```python
df.to_csv('Imgs.csv', index = False)
```
